```python
from bs4 import BeautifulSoup
import urllib2, sys, random, time
```

```python
gameurl = "http://indierecommender.net/output?game_select=007+Legends"
d = urllib2.urlopen(gameurl).read()
soup = BeautifulSoup(d)
```

```python
soup.prettify
```

```python
#a = soup.find("table", {"class": "table-hover"})
table = soup.find('table', 'table table-hover')
links = table.findAll('a')
for link in links:
    print link['href']
```

```python
mylinks = [link['href'] for link in links]
mylinks
```

```python
gameurl = mylinks[0]
z = urllib2.urlopen(gameurl).read()
soup_game = BeautifulSoup(z)
```

```python
mydivs = soup_game.findAll("div", { "class" : "row clear" })
```

```python
mylist = []
for a in mydivs:
    mylist.append(a.text)
```

```python
for l in mylist:
    a = l.split("\n")
    if 'Platform' in a or 'Platforms' in a:
        platform = a[2]
    if 'Engine' in a:
        engine =  a[2]
        engine = engine.encode('utf-8')
    if 'Release Date' in a:
        release_date = a[2]
        release_date = release_date.replace(",", "")
    if 'Genre' in a:
        genre = a[2]
    if 'Theme' in a:
        theme =  a[2]
    if 'Players' in a:
        players = a[2]
        players = players.replace(",", "")
```

```python
platform, engine, release_date, genre, theme, players
```

```python
divScore = soup_game.findAll("div", { "class" : "score" })
divRating = soup_game.findAll("div", { "class" : "rating" })
score = []
rating = []
for s in divScore:
    score.append(s.text)
for r in divRating:
    rating.append(r.text)

s = score[0].replace('\n', '')
score_final = s.split("Average")[-1]
rating = (rating[0].strip("\n"))[:2]
```

```python
score_final, rating
```
