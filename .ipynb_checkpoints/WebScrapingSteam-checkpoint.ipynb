```python
from bs4 import BeautifulSoup
import urllib2
r = urllib2.urlopen("http://www.indiedb.com/groups/2010-indie-of-the-year-awards/top100").read()
soup = BeautifulSoup(r)
```

```python
print soup.prettify()
```

```python
for a in soup.find_all('a', href=True):
    print a['href']
```

```python
games = soup.find_all('h4')
```

```python
for game in games:
    print game.text
```

```python
with open('Indie2010.txt', 'w') as f:
    for game in games:
        f.write(game.text.encode('utf8')+"\n")
f.close()
```
